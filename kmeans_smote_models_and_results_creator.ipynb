```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import random
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report , accuracy_score
from sklearn.model_selection import train_test_split
import math
import csv
```

```python
def pipeline(no_of_features , all_features , x_train , x_test , y_train , y_test , n_estimators = 100 , max_depth=None):
  features = all_features.copy()
  random.shuffle(features)

  selected_features = features[:no_of_features]
  x_train = x_train[selected_features]
  x_test = x_test[selected_features]

  classifier = RandomForestClassifier(n_estimators=n_estimators , max_depth = max_depth , random_state=42)
  classifier.fit(x_train,y_train)
  y_test_pred = classifier.predict(x_test)
  y_train_pred = classifier.predict(x_train)

  return accuracy_score(y_test,y_test_pred) , accuracy_score(y_train,y_train_pred) , selected_features
```

```python

import pandas as pd
end = "2_3"
# Read the CSV file
tryi = pd.read_csv(f'kmeans_smote_results/results_{end}.csv')

# Filter rows where the difference between test and train accuracy is less than 0.05 (no underfitting)
filtered_df = tryi[(tryi['test_accuracy'] - tryi['train_accuracy']) < 0.05 ]

# From these, select the row(s) with the maximum test accuracy
max_test_acc_row = filtered_df[filtered_df['test_accuracy'] == filtered_df['test_accuracy'].max()]
# max_test_acc_row = filtered_df[filtered_df['test_accuracy'] > 0.9]
# max_test_acc_row = max_test_acc_row[filtered_df['train_accuracy'] > 0.999]

max_test_acc_row
```

```python
# Best selected features from previous results
selected_features = ['slope', 'restecg', 'fbs', 'trestbps']
# Read data
x_train = pd.read_csv(f"kmeans_smote_dataset/X_train_scaled_{end}.csv")[selected_features]
y_train = pd.read_csv(f"kmeans_smote_dataset/y_train_{end}.csv")
x_test = pd.read_csv(f"kmeans_smote_dataset/X_test_scaled_{end}.csv")[selected_features]
y_test = pd.read_csv(f"kmeans_smote_dataset/y_test_{end}.csv")

# Convert y from DataFrame to Series (to fix warning)
y_train = y_train.values.ravel()
y_test = y_test.values.ravel()

# Train and evaluate
classifier = RandomForestClassifier(n_estimators=5, max_depth=5, random_state=42)
classifier.fit(x_train, y_train)

y_test_pred = classifier.predict(x_test)
y_train_pred = classifier.predict(x_train)

print("testing")
print(classification_report(y_test, y_test_pred,digits=6))
print("training")
print(classification_report(y_train, y_train_pred , digits=6))

```
